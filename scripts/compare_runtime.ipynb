```python
import matplotlib.pyplot as plt
import matplotlib as mpl
elapsed_time = [94, 41+4*60, 43*60, (20*60+28)*60, 280, 89, 54]
methods = ['T1SEstacker', 'T3SEpp', 'CNN-T4SE', 'Bastion6', 'EffectiveDB', 'DeepSecE', 'MoCETSE']

mpl.rcParams['font.family'] = 'Arial'
mpl.rcParams['axes.labelsize'] = 12
mpl.rcParams['axes.titlesize'] = 14
mpl.rcParams['legend.title_fontsize'] = 12
mpl.rcParams['legend.fontsize'] = 12
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

elapsed_time_ = [seconds/60 for seconds in elapsed_time]

plt.figure(figsize=(8, 5))
plt.barh(methods, elapsed_time_, height=0.6, color='#7dd0b6') 

plt.xscale('log') 
plt.xlim([0.5, 24*60])
plt.xticks([1, 10, 100, 1000], labels=['1', '10', '100', '1000'])

vertical_lines = [1, 10, 100]
for line in vertical_lines:
    plt.axvline(x=line, color='black', alpha=0.3, linestyle='--')
    
plt.xlabel('Elapsed time (min)')
plt.ylabel('Method')
plt.title('Prediction on Dataset S2 (260 sequences)')

sns.despine()
plt.tight_layout()
plt.savefig('png/comparison_runtime_horizontal.png', dpi=600, bbox_inches='tight')
```
